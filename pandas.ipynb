-> if two series have some common elements, then mapping two of them together makes it a dataframe where the common elements are joined with the label name and the ones without any label take value as 'NONE'

- Numpy - high performanc, array computing ideas 
PANDAS COMBINES THE ABOVE BOTH

# DataFrame

--> Data frame is a collection of columns in a rectangular tabular data. could be numeric, string, boolean etc. 

--> The data frame has both row and column index. It can be thought of as dict of series all sharing same index. 

--> The data underneath is stored as a two- dimensional block rather than a one_D array(list, dict etc)

    > To create dataframe:
    1. from a dict of equal length, tuple with all values of row together
    2. numpy arrays
