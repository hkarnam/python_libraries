--> We use visualizations or graphs to give a pictorial representation of our data which makes it easier to understand and analyse

--> What is our data? what type is it? 
--> What do we need to convey through the graph? What is the question that we need to answer? 

Our data can be divided into two types:
1. Quantitative - numerical variables - You can perform arithmetic operations on the data. 
  ex: students in class, heights, weights, temperature, population count etc, 
  
  1. discrete variable - count - where the count is finite. 
          ex: runs scored in match, no of students in class, vote count, covid cases etc, 
  2. Continous variable - infinite values - where count is limited by our capability and willingness to calculate/           measure. 
          ex: Pi values is continous, height, weight, temperature, percentage of something(with decimal points)
          - A continous may take up finite value in some cases, but intrinsicly not possible ( can be different              for various reasons( people, instrumnets)- VALUES CANNOT BE SUBJECTED TO A COUNT THOUGH. 
    
2. Qualitative - categorical data - grouping can be done 
  ex: gender, citizenship, education level etc. 
  
  1. Ordinal - where variables have some order to it. 
     ex: feedback rating - stars 1 to 5 - we know that 5 star is better than 1 star. 
     -socia economic status - low income, middle class, high income
     -education level - high school, grad, post grad, phd
   
  2. Nominal- where the variables have no order at all
     ex: race, gender, nationality etc, 
    
  -->  Nominal - when they have only two possibilites, ex: yes/no, male/female, black/white etc. 
  
  
How to choose a visualization? 4 categories: 

1. composition charts - pie charts,stat bar, line graphs - with shaded AUC - tells how much is a part of data in whole composition. 

2. Distribution chart - different segments of data, how is my data distributed from time to time or class to other class, country population in world
   -- Focusing question: How are the datapoints behaving with respect to statistical calculations such as mean, median, mode, standard deviation etc?

3. Relationship chart - see the relationship between two variables. ex: as salary increase, job satisfaction increases or vice versa 

4. comparision charts - trying to compare variables - ( could also be two line charts against each other, count chart/ histogram)

# Histograms: 

Histogram- a graph- that organizes datapoints --- into specified ranges. (DISTRIBUTION CHART)

--> It takes a maximum and minimum value and dvides the difference into appropriate number of bins (can be done manually or default in matplotlib)
--> and places each datapoint in appropriate bin( based on the range they have) 

--> Each category/bar is called a bin. By default, the data is divided into 10 bins.( this can be changed later by using bins parameter)


```python
firstx = np.random.randint(1,101,200)
firstx
```

```python
plt.figure(figsize =(16,10))
plt.hist(firstx)
plt.show()
```

```python
plt.figure(figsize =(16,10))
plt.hist(firstx,rwidth=0.9,bins = [10,30,70,100])  
# specifying number of bins with values - hist provides bars based on range from one number to other
plt.show()
```

```python
plt.figure(figsize =(16,10))
plt.hist(firstx,rwidth=0.9, range = (30,80), bins = 15)  # can specify only certain values with diff num of bins
plt.show()
```

# adjusting the spacing around subplots: 

The spacing between the subplots are managed automatically by matplotlib relative to the height and width of figure object we specify in the beginnig. The plot will adjust dynamically itsel, if we resize the plot either programatically or manually. 

--> otherways we have subplots_adjust to control the spaces in between and around subplots 

```python
fig,axes = plt.subplots(2,2,figsize = (14,6), sharex = True, sharey = True)
```

```python
fig,axes = plt.subplots(2,2,figsize = (14,6), sharex = True, sharey = True)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.randint(0,500,100), bins = 50, alpha = 0.5) # generating random histogram 4 times
plt.subplots_adjust(wspace = 0.0, hspace = 0.0)
plt.show()
```

```python
fig,axes = plt.subplots(2,2,figsize = (14,6), sharex = True, sharey = True)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.randint(0,500,100), bins = 50, alpha = 0.5) # generating random histogram 4 times
plt.subplots_adjust(wspace = 0.5, hspace = 0.5) # whatever the width was reduce it by 0.5
plt.show()
```
